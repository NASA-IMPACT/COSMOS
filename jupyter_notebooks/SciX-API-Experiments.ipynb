```python
encoded_query = urlencode({"q": "mouse liver",
                           "fl": "title, bibcode, links_data, author, abstract",
                           "rows": 2000,
                           "sort": "score desc"
                          })
results = requests.get("https://scixplorer.org/v1/search/query?{}".format(encoded_query), \
                       headers={'Authorization': 'Bearer ' + token}).json()
results
```
